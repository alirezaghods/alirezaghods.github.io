```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
figure, axes = plt.subplots()
draw_circle1 = plt.Circle((0.5, 0.5), 0.3, color='lawngreen')
draw_circle2 = plt.Circle((0.5, 0.5), 0.2, color='w')
draw_circle3 = plt.Circle((0.5, 0.5), 0.1, color='r')
axes.set_aspect(1)
axes.add_artist(draw_circle1)
axes.add_artist(draw_circle2)
axes.add_artist(draw_circle3)
plt.axis('off')
plt.savefig('logo.png', transparent=True, bbox_inches='tight', pad_inches = 0)
```
